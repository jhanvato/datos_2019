```python
# importacion general de librerias y de visualizacion (matplotlib y seaborn)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('default') # haciendo los graficos un poco mas bonitos en matplotlib
#plt.rcParams['figure.figsize'] = (20, 10)

sns.set(style="whitegrid") # seteando tipo de grid en seaborn
```

```python

clicks['created']=clicks['created'].astype('datetime64[ns]');
clicks.head()
```

```python
clicks['dia']=pd.DatetimeIndex(clicks['created']).day;
clicks['mes']=pd.DatetimeIndex(clicks['created']).month;
clicks['semana']=pd.DatetimeIndex(clicks['created']).week;
clicks['diadelasemana']=pd.DatetimeIndex(clicks['created']).weekday;
clicks['hora']=pd.DatetimeIndex(clicks['created']).hour;
dias_de_la_semana={0:'Lunes',1:'Martes',2:'Miercoles',3:'Jueves',4:'Viernes',5:'Sabado',6:'Domingo'}
clicks['diadelasemana'].replace(dias_de_la_semana, inplace=True)
clicks.head()
```


# Analisis del archivo Clicks

```python
g = sns.countplot(x=clicks['diadelasemana'], orient='v', order=['Lunes','Martes','Miercoles','Jueves','Viernes','Sabado','Domingo'], palette=['blue'])
g.set_title("clicks por dÃ­a de la semana", fontsize=18)
g.set_xlabel("DÃ­a",fontsize=18)
g.set_ylabel("Cantidad de clicks", fontsize=18)
```

```python
g = sns.countplot(x=clicks['dia'], orient='v', palette=['blue'])
g.set_title("clicks por dÃ­as", fontsize=18)
g.set_xlabel("DÃ­a",fontsize=18)
g.set_ylabel("Cantidad de clicks", fontsize=18)
```

```python
categories = list(clicks.groupby("hora").count().index)
# number of variable
N = len(categories)

from math import pi
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

# Initialise the spider plot
ax = plt.subplot(111, polar=True)
 
# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels labels yet
plt.xticks(angles[:-1], categories)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([500,1000,1500,2000,2500,3000], ["500","1000","1500","2000","2500","3000"], color="grey", size=10)
plt.ylim(0,3500)
 

# Ind2
#values = [10,15,20,13,5]
values = list(clicks.groupby("hora")["hora"].count())
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Cantidad de clicks por Hora")
ax.fill(angles, values, 'b', alpha=0.1)

plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
```

```python
diadelasemanaxhora = clicks.groupby(["diadelasemana","hora"])["created",].count()
diadelasemanaxhora = diadelasemanaxhora.reset_index()
pivot_post = diadelasemanaxhora.pivot_table(index= "hora", columns="diadelasemana", values= "created")
pivot_post = pivot_post.sort_index(ascending=False)
plt.figure(figsize=(7,5.6))
cmap = sns.cm.rocket_r
#plt.set_xticklabels(['Mon','Tue','Wen','Thu','Fri','Sut','Sun'])
#plt.xthic(["Lu","Ma","Mi","Ju","Vi","Sa","Do"])
g = sns.heatmap(pivot_post, xticklabels = ["Lu","Ma","Mi","Ju","Vi","Sa","Do"] ,cmap= cmap)
g.set_title("Dia de la semana x hora de clicks")
```

```python
dia = clicks['dia'].value_counts().sort_index().reset_index()
dia.rename(columns={'index':'Dia','dia':'cantidad_de_clicks'},inplace=True)
dia.set_index('Dia',inplace=True)
dia
```

```python
graf=dia.plot(kind='line',figsize=(5,5),fontsize=12)
graf.set_title("Cantidad de clicks por Dia",fontsize=20)
graf.set_xlabel("Dia",fontsize=12)
graf.set_ylabel("Cantidad de clicks",fontsize=12)
```
